```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# 必要なライブラリをインストール


# ログイン情報
login_url = "http://10.226.10.57:81/Moulding"
username = "200000"
password = "200000"

# セッションを作成してログインページにアクセス
session = requests.Session()
login_page = session.get(login_url)

# ログインフォームのデータを作成
soup = BeautifulSoup(login_page.content, "html.parser")
form = soup.find("form")

# __RequestVerificationTokenを取得
verification_token_input = form.find("input", {"name": "__RequestVerificationToken"})

# __RequestVerificationTokenが見つかったかどうかを確認
if verification_token_input is not None:
    verification_token = verification_token_input["value"]

    # ログインデータ
    payload = {
        "staff_id": "200000",
        "password": "200000",
        "submit": "login",
        "__RequestVerificationToken": verification_token
    }

    # ログインを試みる
    login_result = session.post(login_url, data=payload)

# ChromeのWebDriverのパスを取得
driver_path = ChromeDriverManager().install()

# WebDriverのオプションを設定
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")  # ヘッドレスモードで起動
chrome_options.add_argument("--disable-gpu")  # GPUを使用しない
chrome_options.add_argument("--no-sandbox")  # セキュリティ設定
chrome_options.add_argument("--disable-dev-shm-usage")  # メモリ不足対策

# WebDriverを起動
driver = webdriver.Chrome(service=Service(driver_path), options=chrome_options)

# ページにアクセス
driver.get("http://10.226.10.57:81/Moulding")

# 日付と号機の要素を取得
date_from = driver.find_element_by_id("DateFrom")
date_to = driver.find_element_by_id("DateTo")
imm_id = driver.find_element_by_id("IMM_ID")
search_button = driver.find_element_by_xpath("//input[@value='検索']")

# 日付を設定
date_from.clear()
date_from.send_keys("2022-01-01")
date_to.clear()
date_to.send_keys("2023-02-05")

# 号機を選択（例：A01）
imm_id.send_keys("A01")

# チェックボックスを選択
del_flg = driver.find_element_by_id("DelFlg")
del_flg.click()

manual_flg = driver.find_element_by_id("ManualFlg")
manual_flg.click()

# 検索ボタンをクリック
search_button.click()

# 2秒待機して結果を確認（適宜時間を調整してください）
time.sleep(2)

# 結果を取得して表示
print(driver.page_source)

# WebDriverを終了
driver.quit()
```
